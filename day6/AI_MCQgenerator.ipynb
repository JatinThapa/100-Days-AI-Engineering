```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_contents
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
contents = fetch_website_contents("https://edwarddonner.com")
contents[:100]
```

```python
system_prompt = """
You are a university professor who is to prepare questions on a given topic.
The topic can be a technical or a non-technical one.
Generate the quesions in a multiple choice manner. Each question should have 4 options with 1 correct, for example:

    Questions:
    Q1. What is a prominent hashfunction websites used to hash user passwords?
    
    A. sha-256
    B. SHA-512
    C. MD5
    D. bcrypt

The question and options should strictly be displayed in separate lines, exactly like the example above.
Generate atleast 10 questions like these, unless the user mentions otherwise.
At the end of all the questions, prepare an answer key that maps the question to the correct option
example:
    Answer key:
    1. D
Please try to include a mix of easy, medium and hard questions.
Reply in markdown
"""
```

```python
def get_user_prompt(url,contents):
    user_prompt = f"\nThe contents of this website {url} is as follows; Please prepare some multiple choice questions for the main topic you see on the page\n"
    user_prompt+=contents;

    return user_prompt;
```

```python
def messages_for(url,contents):
    return [{"role":"system","content":system_prompt},
            {"role":"user","content":get_user_prompt(url,contents)}]
```

```python
def summarize(url):
    contents = fetch_website_contents(url)
    ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
    stream = ollama.chat.completions.create(
        model=MODEL_LLAMA,
        messages=messages_for(url,contents),
        stream=True
    )

    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
def display_summary(url):
    summary = summarize(url);
    display(Markdown(summary))
```
